# Splitting

```python
# Removing irrelevant folder.
%cd human-seg-dataset/
!rm -rf matting_human_half
%cd ../
```

```python
file_paths_img = []
for root, dirs, files in os.walk("/content/human-seg-dataset/clip_img/"):
  if files:
    for file in files:
      p = os.path.join(root, file)
      file_paths_img.append(p)
```

```python
file_paths_label = []
for root, dirs, files in os.walk("/content/human-seg-dataset/matting/"):
  if files:
    for file in files:
      p = os.path.join(root, file)
      file_paths_label.append(p)
```

```python
len(file_paths_img), len(file_paths_label)
```

```python
file_paths_img.sort()
file_paths_label.sort()
```

```python
splits = {}
num_splits = 4
num_instances = len(file_paths_img)
for i in range(num_splits):
  splits[f"split_{i}"] = {}
  splits[f"split_{i}"]["img_paths"] = file_paths_img[(i * num_instances) // num_splits:((i + 1) * num_instances) // num_splits]
  splits[f"split_{i}"]["label_paths"] = file_paths_label[(i * num_instances) // num_splits:((i + 1) * num_instances) // num_splits]
```

```python
for i in range(num_splits):
  print(f"split_{i} count:", len(splits[f"split_{i}"]["img_paths"]), len(splits[f"split_{i}"]["label_paths"]))
```

# Saving Zip Files

```python
"/".join(splits[f"split_{i}"]["img_paths"][0].split("/")[3:])
```

```python
for i in range(num_splits):
  with ZipFile(f"/content/split_{i}.zip", "w") as f:
    for img_p, label_p in zip(splits[f"split_{i}"]["img_paths"], splits[f"split_{i}"]["label_paths"]):
      f.write(img_p, arcname=img_p.split("/")[-1])
      f.write(label_p, arcname="label_" + img_p.split("/")[-1])
```
